```python
#Create a function for return the remaining service years return
def get_remaining_service(row):
    retire_age = 75
    service_age = int(row['age'])    
    remaining_service = retire_age - service_age    
   
    return remaining_service
```

```python
#Create a new data set with columns Age, length_of_service, gender_full 
editdataset = data[["age","length_of_service","gender_full"]].sample(20)
editdataset
```

```python
# Create a new column "rem_service_year" and Call the function "get_remaining_service" for the value
editdataset['rem_service_year'] = editdataset.apply(get_remaining_service, axis=1)

#Sort the dataset values
editdataset.sort_values('rem_service_year', ascending=True).head(20)
```
