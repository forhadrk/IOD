```python
# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 17, 35, 29]

# Define figure size and style
plt.figure(figsize=(8, 6)) # Bar Width Size
plt.style.use('ggplot')  # Using a built-in style for a clean look

# Create bar chart
bars = plt.bar(categories, values, color=['#2E86C1', '#A93226', '#239B56', '#F1C40F'], edgecolor='gray', linewidth=1.5)

# Add title and labels
plt.title('Customized Bar Chart Example', fontsize=16, fontweight='bold', color='#2C3E50')
plt.xlabel('Categories', fontsize=12, color='#34495E')
plt.ylabel('Values', fontsize=12, color='#34495E')

# Customize ticks
plt.xticks(fontsize=10, rotation=70, color='#34495E')  # Rotating for better readability
plt.yticks(fontsize=10, color='#34495E')

# Adding values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points", ha='center', va='bottom', fontsize=10,fontweight='bold', color='#2C3E50')

# Add space at the top
plt.subplots_adjust(top=0.60)  # Adjust the top margin (1.0 is max)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
N = 5
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
# the x locations for the groups
ind = np.arange(N)    
# the width of the bars
width = 0.35      

p1 = plt.bar(ind, menMeans, width, yerr=menStd, color='red')
p2 = plt.bar(ind, womenMeans, width,
bottom=menMeans, yerr=womenStd, color='green')

plt.ylabel('Scores')
plt.xlabel('Groups')
plt.title('Scores by group\n' + 'and gender')
plt.xticks(ind, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

```python
# Sample data
A = [1, 2, 3, 4, 5]
B = [5, 4, 3, 2, 1]

# Define colors for each point
colors = ['r', 'g', 'b', 'orange', 'purple']  # One color for each pair (A, B)

# Scatter plot with different colors
plt.scatter(A, B, c=colors, s=100)  # 'c' for color, 's' for size (each is passed once)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot with Different Colors")

plt.show()
```

```python
# Generate points with dynamic colors
# Sample data
A = [1, 2, 3, 4, 5]
B = [5, 4, 3, 2, 1]

# Dynamically generate colors based on values in A (you can also use B or other arrays)
colors = A  # Values used to map colors

# Scatter plot with dynamically assigned colors
plt.scatter(A, B, c=colors, cmap='viridis', s=50)  # 'viridis' is a colormap
plt.colorbar()  # Adds a color bar to show the mapping of colors to values

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot with Dynamic Colors")

plt.show()
```

```python
# create random data
no_of_balls = 25
x = [random.triangular() for i in range(no_of_balls)]
y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
colors = [random.randint(1, 4) for i in range(no_of_balls)]
areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]
# draw the plot
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
